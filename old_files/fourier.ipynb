```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from utility import *
from fourier import *
```

```python
image_path = 'image.jpg'
img, img_denoised, img_bin, th = pre_processing(image_path)
contours = detect_contour(img_bin)

contour_image = np.zeros_like(img)

gx,gy = sample_polygon_uniformly(contours,200)
contour = np.column_stack((gx, gy)).astype(np.int32)
```

```python
original_shape = img.shape
canvas = np.zeros(original_shape)
cv2.drawContours(canvas, [contour], 0, (255), thickness=1)
plt.imshow(canvas, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')
plt.show()
```

```python
f_descriptors_original = fourier_descriptors(contour)

num_coefficients = 50
f_descriptors_reduced =  reduce_fourier_length(f_descriptors_original, num_coefficients)

original_shape = img.shape

reconstructed_image = reconstruct_image(f_descriptors_reduced/(len(f_descriptors_original)/len(f_descriptors_reduced)), original_shape)

plt.imshow(reconstructed_image, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')
plt.show()
```
